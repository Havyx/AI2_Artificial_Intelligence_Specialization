# Exemplo de probabilidade

### Probabilidade de Eventos Dependentes
### Dado o nascimento de duas crianças:
#### * Qual a probabilidade de nascer duas meninas, sendo a mais velha uma menina (1/4 / 2/4) = 1/4 * 4/2 => 0,5
#### * Qual a probabilidade de nascer duas meninas, sendo qualquer das crianças menina (1/4 / 3/4) = 1/4 * 4/3 => 0,3

```python
#probabilidade de eventos dependentes
both_girls = 0
older_girl = 0
either_girl = 0
random.seed(random.random())

for _ in range(10000):
    younger = random.choice(["boy", "girl"])
    older = random.choice(["boy", "girl"])
    if older == "girl":
        older_girl += 1
    if older == "girl" and younger == "girl":
        both_girls += 1
    if older == "girl" or younger == "girl":
        either_girl += 1

print ("P(both | older):", both_girls / older_girl)
print ("P(both | either): ", both_girls / either_girl)
```

## Probabilidade após jogar um dados de ocorrer coroa 0, 1 ou 2 vezes

```python
y=[0.25, 0.5, 0.25] #probabilidade associada
x=[0, 1, 2] #evento

npa = np.asarray(y, dtype=np.float32)
print(npa.sum()) # soma da probabilidade de todos os eventos possíveis = 1

#Criando 3 axes - um para cada gráfico
#fig, ax_lst = plt.subplots(1, 2)
plt.plot(x,y,'ro')

plt.suptitle('Distribuição de probabilidade')

plt.show()
```

```python
op1=0
op2=0
op3=0

t = np.arange(0,1000)


for x in range(1000):
    ht1 = random.choice(["0", "1"])
    ht2 = random.choice(["0", "1"])
    if (ht1 == "0") & (ht2 == "0"):
        op1 += 1
        t[x]=0
    if (ht1 == "0") & (ht2 == "1"):
        op2 += 1
        t[x]=1
    if (ht1 == "1") & (ht2 == "0"):
        op2 += 1
        t[x]=1
    if (ht1 == "1") & (ht2 == "1"):
        op3 += 1
        t[x]=2
    

print(op1,op2,op3)
total=(op1+op2+op3)/1000
print(total)

y=[op1, op2, op3] #probabilidade associada
x=[0, 1, 2] #evento

npa = np.asarray(y, dtype=np.float32)
print(npa.sum()) # soma da probabilidade de todos os eventos possíveis = 1

#Criando 3 axes - um para cada gráfico
#fig, ax_lst = plt.subplots(1, 2)
plt.plot(x,y,'ro')

plt.suptitle('Distribuição de probabilidade')

plt.show()
```

## Distribuição Normal

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.default_rng().normal(mu, sigma, 1000)

fig, ax_lst = plt.subplots(2, 1)

plt.ylabel('Frequência da Distribuição')

count, bins, ignored = ax_lst[0].hist(s, 30) 

ax_lst[1].plot(count)

plt.show()
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.default_rng().normal(mu, sigma, 1000)

count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')

plt.show()
```

## Testes estatísticos

## Verificando hipótese de normalidade

```python
from scipy import stats

x = stats.norm.rvs(loc=5, scale=3, size=100)
stat, pvalor = stats.shapiro(x)
print('Statistica=%.10f, p=%.10f' % (stat, pvalor))

alpha = 0.05

if pvalor > alpha:
    print('Similar a curva Gaussiana (falha para rejeitar H0)')
else:
    print('Não é similar a curva Gaussiana (rejeita H0)')
```

```python
from scipy.stats import shapiro
stat, pvalor = shapiro(height['HEIGHT'])

stat, pvalor = shapiro(height['HEIGHT'])
print('Statistica=%.10f, p=%.10f' % (stat, pvalor))

alpha = 0.05

if pvalor > alpha:
    print('Similar a curva Gaussiana (falha para rejeitar H0)')
else:
    print('Não é similar a curva Gaussiana (rejeita H0)')
```
