```python
values= np.random.normal(80,2, 1000)
df = pd.DataFrame(values)
print(df.head())
print(df.describe())
print("Mediana")
print(df.median())
df.hist()
df.plot.density()
df.plot.box()
```

# Distribuição Normal e Quartil

```python
meanaux=df.mean()
mean=meanaux[0]

stdaux=df.std()
std=stdaux[0]

print(mean-(2*std))
print(mean-(std))
print(mean+(std))
print(mean+(2*std))


df68=df.loc[(df[0] >= (mean-std)) & (df[0] <=  (mean+std))]

print(df68.describe())

df95=df.loc[(df[0] >= (mean-(2*std))) & (df[0] <=  (mean+(2*std)))]

print(df95.describe())

df26=df.loc[(((df[0] >= (mean-(2*std))) & (df[0] <= (mean-(std)))) | (((df[0] <=  (mean+(2*std)))) & (df[0] >=  (mean+(std)))))]

print(df26.describe())
```

```python
#Marcando quartis
bin_labels_5 = ['1', '2', '3', '4','5','6']
df['quartile'] = pd.qcut(df[0], q=[0, 0.05, .25, .5, .75, .95, 1], labels=bin_labels_5)

df2=df.loc[df['quartile'] == '6']
print(df2.describe())
```

```python
from scipy.stats import zscore

df3 = df.apply(zscore) 
print(df3)
print(df3.describe())
```
