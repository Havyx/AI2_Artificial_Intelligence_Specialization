## Abra a imagem figs/messi5.jpg

```python
img = cv.imread("./figs/messi5.jpg")
figure(figsize=(10,10))
plt.imshow(img)
```

## Recorte a bola

```python
figure(figsize=(10,10))
bola = img[284:390, 335:390]

plt.imshow(bola)
```

## Copie em outra parte da imagem

```python
figure(figsize=(10,10))
bola = img[284:390, 335:390]
cabeca = img[71:129, 210:265]

trocaImage = img.copy()
trocaImage[71:129,  210:265] = bola
trocaImage[284:390, 335:390] = cabeca

plt.imshow(trocaImage)
```

## Rotacione a imagem em 60 graus

```python
figure(figsize=(10,10))

trocaImage = img.copy()
trocaImage[71:129,  210:265] = imutils.rotate(bola, 45)
trocaImage[284:390, 335:390] = imutils.rotate(cabeca, 60)

plt.imshow(trocaImage)
```

## Altere o tamanho da imagem mantendo a proporção

```python
figure(figsize=(10,10))

multiplicador = 5
newImage = imutils.resize(trocaImage, width=(trocaImage.shape[0] * multiplicador))
plt.imshow(newImage)
```
