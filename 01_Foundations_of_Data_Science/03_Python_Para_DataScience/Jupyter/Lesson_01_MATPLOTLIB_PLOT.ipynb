```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
z= x**3
```

```python
x
```

```python
#Fuction
plt.plot(x,y, 'r-')
#plt.show()
plt.xlabel('XLabel')
plt.ylabel('Ylabel')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
#OBJECT ORIENTED
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X_LABEL')
axes.set_ylabel('Y_LABEL')
axes.set_title('TITLE')
```

```python
fig2 = plt.figure()

axes1 = fig2.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig2.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x)
axes2.set_title('TITLE')
axes1.set_title('TITLE LARGER')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(x,y)
axes[1].set_xlabel('TEESTE')
```

```python
fig = plt.figure(figsize=(8,2), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4), dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```
